# Que 1

Que 1: Let there be two matrices A(n,k) and B(k,m). The multiplication of these matrices will generate a new matrix with dimension(n,m), which means there are ( n x m ) elements in it. Each element is obtained by its particular row's dot multiplication to its column. While the dot product occurs there are k times multiplication and (k-1) times addition to obtain a single element of a new matrix. Since there are (n x m) elements hence the number of multiplications are (n x m x k) and the number of additions are (n x m x (k - 1)).

```python
## Que 2:
import time
C=[]
A= [[1,2,3],[3,4,5]]
B= [[1,2,4,5],[3,4,6,7],[5,6,9,8]]
def mat(A, B):
    ans = [[0 for i in range(len(B[0]))] for i in range(len(A))]

    for i in range(len(A)):
        for j in range(len(B[0])):
            for k in range(len(B)):
                ans[i][j] += A[i][k] * B[k][j]
    return ans
start = time.time()
result = mat(A,B)
print(result)
end = time.time()
print(end - start)
```

```python
import numpy as np
A= [[1,2,3],[3,4,5]]
B= [[1,2,4,5],[3,4,6,7],[5,6,9,8]]
start1 = time.time()
np.matmul(A,B)
end1 = time.time()
print(end1 - start1)
print(np.matmul(A,B))
```

The numpy matrix multiplication is much faster than the function defined for multiplication using lists of list. This is because the numpy library is written in C language which is faster than python. Also it devides the task into subtasks and do it parallely.

```python
## Que 3:

## using the 2 passes of the array
def bubble_sort(arr):
  n = len(arr)
  for i in range(n):
          for j in range(0, n - i - 1):
              if arr[j] > arr[j + 1]:
                  arr[j], arr[j + 1] = arr[j + 1], arr[j]

S = [1, 8, 0, 2, 4,4]
n = len(S)
bubble_sort(S)
if n % 2 == 0:
    print((S[int((n/2)-1)]+S[int((n/2))])/2)
else:
    print(S[int(n/2)])

for i in range(len(S)):
    print(S[i],end =" ")
## the time complexity of this is O(n**2)

# the time complexity can be reduced by using quick sort to find the median, the quick sort has time complexity as O(nlog(n))
```

```python
## Ques 5
# Q: Use JAX to confirm the gradient evaluated by your method matches the analytical solution corresponding to a few random values of x and y
import jax.numpy as jnp
from jax import grad

def f(x, y):
    return x**2*y + y**3*jnp.sin(x)

grad_f_x = grad(f, argnums=0)
grad_f_y = grad(f, argnums=1)

x = jnp.array(5.0)
y = jnp.array(7.0)

grad_x = grad_f_x(x, y)
grad_y = grad_f_y(x, y)

print("Gradient with respect to x:", grad_x)
print("Gradient with respect to y:", grad_y)
```

```python
class Student():
    def __init__(self,roll_number, name, marks):
        self.name = name
        self.roll_number = roll_number
        self.marks = marks
    def __repr__(self):
        return f"Roll_number: {self.roll_number}, Name: {self.name}, Marks: {self.marks})"

class Branch():
    def __init__(self,branch):
        self.branch = branch
        self.students = []
    def add_student(self,student):
        self.students.append(student)
    def __repr__(self):
        return f"{self.branch}: {self.students})"
class Year():
    def __init__(self,year):
        self.year = year
        self.branches = []
    def add_branch(self,branch):
        self.branches.append(branch)
    def __repr__(self):
        return f"{self.year}:{self.branches})"
class data():
    def __init__(self):
        self.years = []
    def add_year(self,year):
        self.years.append(year)
    def __repr__(self):
        return f"{self.years}"



student1 = Student("00201", "X", {"Math": 90, "Science": 85})
student2 = Student("78429", "Y", {"Math": 75, "Science": 80})
student3 = Student("12345", "Z", {"Math": 95, "Science": 90})
student4 = Student("67890", "A", {"Math": 85, "Science": 95})

branch1 = Branch("Mechanical")
branch1.add_student(student1)
branch1.add_student(student2)

branch2 = Branch("Chemical")
branch2.add_student(student3)
branch2.add_student(student4)

year1 = Year("2024")
year1.add_branch(branch1)

year2 = Year("2023")
year2.add_branch(branch2)

database = data()
database.add_year(year2)
database.add_year(year1)

print(database)
```

```python
import matplotlib.pyplot as plt

x = np.arange(0.5,100.0,0.5)
y1= x
y2= x**2
y3= x**3/100
y4= np.sin(x)
y5= (np.sin(x))/x
y6= np.log(x)
y7= np.exp(x)

plt.figure(figsize=(15, 15))

plt.subplot(3, 3, 1)
plt.plot(x, y1, "r", label="y = x")
plt.title("y = x")
plt.legend()
plt.grid(True)

plt.subplot(3, 3, 2)
plt.plot(x, y2, "g", label="y = x^2")
plt.title("y = x^2")
plt.legend()
plt.grid(True)

plt.subplot(3, 3, 3)
plt.plot(x, y3, "b", label="y = x^3 / 100")
plt.title("y = x^3 / 100")
plt.legend()
plt.grid(True)

plt.subplot(3, 3, 4)
plt.plot(x, y4, "y", label="y = sin(x)")
plt.title("y = sin(x)")
plt.legend()
plt.grid(True)

plt.subplot(3, 3, 5)
plt.plot(x, y5,label="y = (sin(x)) / x")
plt.title("y = sin(x) / x")
plt.legend()
plt.grid(True)

plt.subplot(3, 3, 6)
plt.plot(x, y6, label="y = log(x)")
plt.title("y = log(x)")
plt.legend()
plt.grid(True)

plt.subplot(3, 3, 7)
plt.plot(x, y7, label="y = exp(x)")
plt.title("y = exp(x)")
plt.legend()
plt.grid(True)

plt.tight_layout()
plt.show()
```
